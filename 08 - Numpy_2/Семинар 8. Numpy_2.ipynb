## Маскирование и специальное индексирование

Маска - массив с элементами типа `bool`.

Маскирование - применение маски к массиву того же размера для выбора только тех элементов массива, которым соответствуют значения `True` в маске. Результат - всегда одномерный массив.

Специальное индексирование - указание в качестве индекса списка или массива из индексов. В массиве индексов элементы могут повторяться. Размер результирующего массива соответствует размеру массива индексов.

Маска и специальное индексирование не копирует часть массива, а создает новое представление `view`, через которое можно как считывать, так и записывать значения элементов части исходного массива.


````python
# массив
arr = np.arange(10)
arr
````

````python
# маска
mask = (arr % 2 == 0)
mask
````

````python
arr[mask]
````

````python
# представление массива для визуализации маски
view = arr.view(type=sp)
view[mask]
````

````python
arr = np.arange(20*20).reshape(-1, 20)
arr
````

````python
mask = (arr > 11) & (arr < 51)
mask
````

````python
arr[mask]
````

````python
# представление массива для визуализации маски
view = arr.view(type=sp)
view[mask]
````

````python
mask = (arr[1] % 2 == 0) | ((arr[1] % 3 != 0))
mask
````

````python
# срез по измерению 0 и маска по измерению 1
arr[1:-1, mask]
````

````python
# представление массива для визуализации маски
view = arr.view(type=sp)
view[1:-1, mask]
````

````python
arr[[1, 2]]
````

````python
view[[1, 2, 10, 13]]
````

````python
arr[:, [0, 2, 4]]
````

````python
view[:, [0, 2, 2, 2, 4]]
````

````python
rows = np.array([1, 3, 7, 13]).reshape(-1, 1)
cols = np.array([0, 2, 5, 8])
arr[rows, cols]
````

````python
view[rows, cols]
````

````python
rows = np.array([1, 3, 4, 1])
cols = np.array([0, 2, 4, 1])
arr[rows, cols]
````

````python
view[rows, cols]
````

## Описательные статистики

- `np.min(a, axis)` - минимум
- `np.max(a, axis)` - максимум
- `np.mean(a, axis)` - среднее
- `np.std(a, axis)` - стандартное отклонение
- `np.var(a, axis)` - дисперсия

````python
rng = np.random.default_rng()
arr = np.int32(rng.uniform(0, 100, size=(10,10)))
arr
````

````python
arr[2, 3]
````

## Работа с размерностью массива

Фиктивной размерностью будем называть размерность массива, которая содержит только один элемент.

Т.е. все единицы в `arr.shape` являются фиктивными размерностями

`np.newaxis` - константа, позволяет добавлять фиктивные размерности в массив; эквивалентна `None`

`arr[:, np.newaxis] или arr[:, None]` - добавить фиктивную размерность сразу после нулевого измерения

`arr[None, :]` - добавление фиктивной размерности в начало, т.е. новая нулевая размерность станет фиктивной

`...` - оператор эллипсис, позволяет выбрать все ранее выбранные размерности массива

`arr[..., None]` - добавить фиктивную размерность в конец списка размерностей

````python
arr = np.arange(5)
arr, arr.shape
````

````python
arr[:, None, None].shape
````

````python
# добавление фиктивной размерности двумя способами
arr[:, None], arr.reshape(-1, 1)
````

````python
arr[None].shape
````

````python
arr[None, ...].shape
````

````python
arr[None, None, :]
````

````python
arr[None, None, :].shape
````

````python
arr = np.zeros((5,5))
arr
````

````python
arr[:, None].shape
````

````python
arr[..., None].shape
````

````python
arr[:, None, :].shape
````
